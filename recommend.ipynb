```python
import ast


def fetchd(x):
    dir = []
    for i in ast.literal_eval(x):
        if i['job'] == 'Director':
            dir.append(i['name'])
    return dir

movies['crew'] = movies['crew'].apply(fetchd)
```

```python
def fetchc(x):
    L = []
    counter =  0
    for i in ast.literal_eval(x):
        if counter < 3:
            L.append(i['name'])
        counter += 1
        
    return L

movies['cast'] = movies['cast'].apply(fetchc)
```

```python
def fetchg(x):
    L=[]
    for i in ast.literal_eval(x):
        L.append(i['name'])
    return L

movies['genres'] = movies['genres'].apply(fetchg)
```

```python
movies['keywords']=movies['keywords'].apply(fetchg)
```

```python
def collapse(x):
    L = []
    for i in x:
        L.append(i.replace(" ",""))
    return L

movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python

movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new_movies = movies.drop(columns=['overview','genres','keywords','cast','crew'])
new_movies.head()
```

```python
new_movies['tags'] = new_movies['tags'].apply(lambda x: " ".join(x))
new_movies.head()
```

## Content Base Recommendation

```python
new_movies = new_movies.drop(columns = 'popularity')
new_movies.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

cv = CountVectorizer(max_features=5000, stop_words = 'english')
```

```python
vectors = cv.fit_transform(new_movies['tags']).toarray()
```

```python
vectors.shape
```

```python
similarity_score = cosine_similarity(vectors)
similarity_score
```

```python
import requests
def fetch_poster(movie_id):
    url = "https://api.themoviedb.org/3/movie/{}?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US".format(movie_id)
    data = requests.get(url)
    data = data.json()
    poster_path = data['poster_path']
    full_path = "https://image.tmdb.org/t/p/w500/" + poster_path
    return full_path
```

```python
fetch_poster(5)
```

```python
def recommend(movie):
    index = new_movies[new_movies['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity_score[index])), reverse=True, key=lambda x: x[1])
    recommended_movie_names = []
    recommended_movie_posters = []
    recommended_movie_release_date = []
    for i in distances[1:6]:
        # fetch the movie poster
        movie_id = new_movies.iloc[i[0]].movie_id
        recommended_movie_posters.append(fetch_poster(movie_id))
        recommended_movie_names.append(new_movies.iloc[i[0]].title)
        recommended_movie_release_date.append(n_movies.iloc[i[0]].release_date)

    return recommended_movie_posters,recommended_movie_names,recommended_movie_release_date
```

```python
recommend('Avatar')
```
